# Problem Statement:

You work for a global retail company that operates in various regions around the world, including the Australia and Oceania region. The company stores sales data in a CSV file named "sales.csv," which contains information about sales and profits in different countries. You are tasked with developing a program to analyze sales data stored in a CSV file named "sales.csv" for countries within the Australia and Oceania region. The program should calculate the maximum, minimum, and average profit for each country in this region and create a bar graph displaying the total profit for each country. This visualization aims to provide insights into the financial performance of countries in the Australia and Oceania region, helping your global retail company make informed decisions.

```python
import pandas as pd
pd.read_csv("sales.csv")
```

```python
#use filter to extract Australia and Oceania
#grouping of total profit for each country
#maximum, minimum, and average profit for each country in this region
#create a bar graph displaying the total profit for each country

import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {
    'Country': ['Australia', 'New Zealand', 'Fiji', 'Papua New Guinea', 'United States', 'Canada'],
    'Region': ['Australia and Oceania', 'Australia and Oceania', 'Australia and Oceania', 'Australia and Oceania', 'North America', 'North America'],
    'Profit': [4400, 3800, 1200, 800, 5000, 4800]
}

df = pd.DataFrame(data)

# Filter to extract Australia and Oceania
oceania_df = df[df['Region'] == 'Australia and Oceania']

# Group by country and calculate total profit
total_profit = oceania_df.groupby('Country')['Profit'].sum().reset_index()

# Calculate maximum, minimum, and average profit for each country
profit_stats = oceania_df.groupby('Country')['Profit'].agg(['max', 'min', 'mean']).reset_index()

# Merge total profit with stats
result = pd.merge(total_profit, profit_stats, on='Country')

print("Total Profit and Statistics by Country:")
print(result)

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(result['Country'], result['Profit'])
plt.title('Total Profit by Country in Australia and Oceania')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the graph
plt.show()
```
